## Inspecting the trend of continuous variables 

```python
continuous_columns = ['LotFrontage', 'LotArea', 'MasVnrArea', 'TotalBsmtSF', 
                      'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', '1stFlrSF', '2ndFlrSF',
                      'LowQualFinSF', 'GrLivArea', 'GarageArea', 'WoodDeckSF',
                      'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch',
                      'MiscVal']
```

```python
def SLR_Calc(X, Y):  # function to calculate slope and intercept
    if isinstance(X, pd.Series):
        X = X.to_frame()
    lm = LinearRegression()
    model = lm.fit(X, Y)
    return(model.coef_[0], model.intercept_)
```

```python
def plot_SLR(X, Y, nrows, ncols, nsubp):  # function to plot data points along with the linear regression of those

    nonzero_ind = X.nonzero()[0]
    
    X = X.iloc[nonzero_ind]
    Y = Y.iloc[nonzero_ind]
    
    X = np.log(X)
    Y = np.log(Y)
    
    plt.subplot(nrows, ncols, nsubp)
    plt.scatter(X, Y, c = 'b')

    slope, intercept = SLR_Calc(X, Y)
    Y_hat = slope * X + intercept

    plt.plot(X, Y_hat, c = 'r');
    plt.title('exponent = '+ str(round(slope, 2)), fontsize=20)
    
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    
    plt.xlabel(X.name, fontsize=20)
    plt.ylabel(Y.name, fontsize=20)
```

```python
plt.figure(figsize=(20, 25))

for i, col in enumerate(continuous_columns, 1):
    plot_SLR(train[col], train['SalePrice'], 6, 3, i)

plt.tight_layout()
```

## Log Transform of Skewed features

```python
numerical_columns = train._get_numeric_data().columns
```

```python
included_numerical_columns = list(set(numerical_columns) - set(excluded_numerical_columns))
```

```python
skewness = train[included_numerical_columns].apply(lambda x: skew(x))

skewness = skewness[abs(skewness) > 0.5]
skewed_features = skewness.index
```

```python
train[skewed_features] = np.log1p(train[skewed_features])
test[skewed_features] = np.log1p(test[skewed_features])
```

## Selection of features

```python
# numerical_columns = train._get_numeric_data().columns
```

```python
# included_numerical_columns = list(set(numerical_columns) - set(excluded_numerical_columns))
```

```python
features = [col for col in included_numerical_columns if col != 'SalePrice']
X = train[features]
y = train['SalePrice']
```

## Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

## Scaling the data

```python
ss = StandardScaler()
ss.fit(X_train)
X_train_sc = ss.transform(X_train)
X_test_sc = ss.transform(X_test)
```

## Elastic Net

```python
l1_ratios = [0.001, 0.005, .01, .05, .1, .15, .2, .3, .5, .7, .9, .95, .99, 1]
elastic = ElasticNetCV(l1_ratio=l1_ratios, n_alphas=1000, max_iter=10000)
```

```python
elastic.fit(X_train_sc, y_train)
```

```python
elastic_optimal_alpha = elastic.alpha_
elastic_optimal_alpha
```

```python
elastic_optimal_l1_ratio = elastic.l1_ratio_
elastic_optimal_l1_ratio
```

```python
elastic_opt = ElasticNet(alpha=elastic_optimal_alpha, l1_ratio=elastic_optimal_l1_ratio, max_iter=10000)
```

```python
elastic_opt.fit(X_train_sc, y_train)
```

```python
elastic_opt.score(X_train_sc, y_train)
```

```python
elastic_opt.score(X_test_sc, y_test)
```

```python
cross_val_score(elastic_opt, X_train_sc, y_train, cv=5).mean()
```

```python
coefficients = pd.Series(elastic_opt.coef_, index = X_train.columns)

print('ElasticNet picked ' + str(sum(coefficients != 0)) + \
      ' features and eliminated the other ' +  str(sum(coefficients == 0)) + ' features.')

imp_coefs = pd.concat([coefficients.sort_values().head(10),
                       coefficients.sort_values().tail(10)])

plt.figure(figsize=(8, 6))
plt.barh(imp_coefs.index, imp_coefs.values)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Coefficients in the ElasticNet Model', fontsize=15);
```

## Analyzing the Residuals

```python
preds = elastic_opt.predict(X_test_sc)
r2_score(y_test, preds)
```

```python
residuals = y_test - preds  
residuals.hist(bins=40);
```
